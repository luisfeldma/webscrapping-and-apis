```python
from pytrends.request import TrendReq
import pandas as pd
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
keywords = ['Trump','Donald Trump', 'Biden', 'Joe Biden']
```

```python
pytrends.build_payload(keywords, timeframe='2017-01-01 2020-10-31')
data = pytrends.interest_over_time()
```

```python
data['Trump Searches'] = data['Trump'] + data['Donald Trump']
data['Biden Searches'] = data['Biden'] + data['Joe Biden']
```

```python
data[['Biden Searches', 'Trump Searches']]['2019':].plot(figsize=(16,8))
```
